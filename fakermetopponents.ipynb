```python
import pandas as pd
```

```python
matchinfo = pd.read_csv('matchinfo__before.csv')
```

```python
columns = ['League', 'Year','Season','Type','gamelength','blueTop','redTop','blueJungle','redJungle','blueTopChamp',
           'blueJungleChamp','blueADC','blueADCChamp','blueSupport','blueSupportChamp','redTopChamp','redJungleChamp','redADC',
          'redADCChamp','redSupport','redSupportChamp','Address','blueMiddleChamp','redMiddleChamp','blueTeamTag','bResult',
          'rResult','redTeamTag']
matchinfo_after_drop = matchinfo.drop(columns, axis=1)
```

```python
#redteam_faker
redteam_faker = (matchinfo_after_drop.redMiddle == 'Faker')
red = matchinfo_after_drop[redteam_faker]
red = red.rename(columns={'redMiddle':'meet_Faker','blueMiddle':'opponents_mid'})
red
```

```python
#blueteam_faker
blueteam_faker = (matchinfo_after_drop.blueMiddle == 'Faker')
blue = matchinfo_after_drop[blueteam_faker]
blue = blue.rename(columns={'blueMiddle':'meet_Faker','redMiddle':'opponents_mid'})
blue
```

```python
df = red.append(blue)
df
```

```python
df = df.groupby('opponents_mid').count()
```

```python
df.to_csv('fakerMetOpponents.csv',index=True) 
```
